```python

def collect_counts(onegene):
    onegenecounts = normcounts[normcounts.index==onegene]
    thesecounts=onegenecounts.T
    IFNBcounts = thesecounts[thesecounts.index.str.contains("IFNB")].copy()
    IFNBcounts.columns = ["IFNB_normcounts"]
    BSAcounts = thesecounts[thesecounts.index.str.contains("BSA")].copy()
    BSAcounts.columns = ["BSA_normcounts"]
    IFNBcounts["newlabel_IFNB"] = IFNBcounts.index
    BSAcounts["newlabel_BSA"] = BSAcounts.index
    thesecounts = paired.merge(BSAcounts, on="newlabel_BSA")
    thesecounts = thesecounts.merge(IFNBcounts, on="newlabel_IFNB")
    return thesecounts


def lin_reg(onegene):
    thesecounts = collect_counts(onegene)
    x=thesecounts["BSA_normcounts"].to_list()
    y=thesecounts["IFNB_normcounts"].to_list()
    x = np.array(x).reshape(-1, 1)
    y = np.array(y)
    reg = LinearRegression().fit(x,y)
    output = [reg.score(x, y), reg.coef_[0], reg.intercept_]
    return output
    
```

```python
onegene="LSM7"
output = lin_reg(onegene)
output
```

```python
outputs = []
for onegene in sig.index.to_list():
    output= lin_reg(onegene)
    outputs.append(output)
    
```

```python
lrdf = pd.DataFrame(outputs, index=sig.index.to_list(), columns = ["R2", "m", "b"])
lrdf
```

```python
lrdf.to_csv(indir+"lrdf.csv")
```
